## Basic Example ResQue Greedy

Example of a basic ResQue Greedy application over a coverage problem.

```python
from RandomCoverageSetting import create_submodular_environment
from Visualizer import plot_environment
from ResQueGreedy import greedy_coverage
from BruteForce import brute_force_optimal_coverage
from Auxiliary import compute_coverage
```

```python
allocation_points, target_points, radius = create_submodular_environment(10, 100, 10, 2, 0.7,0.5)
plot_environment(allocation_points, target_points, radius)
k = 3
```

```python
selected,query_time,rewires,trace_curvature = greedy_coverage(allocation_points, target_points, radius, k)
#plot_environment(allocation_points, target_points, radius, selected)
print("Selected >> ",selected, " using ",rewires," rewires!")
print(f"Total coverage for {k} selected points: {len(compute_coverage(allocation_points[selected], target_points, radius))}")
print(f"Trace curvature is {max(trace_curvature)}")
```

```python
selected_optimal,query_time = brute_force_optimal_coverage(allocation_points, target_points, radius, k)
#plot_environment(allocation_points, target_points, radius, selected_optimal)
print("Selected >> ",selected_optimal)
print(f"Total coverage for {k} selected points: {len(compute_coverage(allocation_points[selected_optimal], target_points, radius))}")
```

```python
selected,query_time,rewires,trace_curvature = greedy_coverage(allocation_points, target_points, radius, k, "ResQue")
#plot_environment(allocation_points, target_points, radius, selected)
print("Selected >> ",selected, " using ",rewires," rewires!")
print(f"Total coverage for {k} selected points: {len(compute_coverage(allocation_points[selected], target_points, radius))}")
print(f"Trace curvature is {max(trace_curvature)}")
```

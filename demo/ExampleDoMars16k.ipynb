# Allocation on Mars exploration

In this example, we apply ResQue Greedy to the DoMars16k.

```python
import numpy as np
from DoMars16k import count_image_keys,generate_random_points
from Auxiliary import compute_coverage
from ResQueGreedy import greedy_coverage
from Visualizer import plot_points
import time
import random
```

```python
base_folder = 'domars16k/'
counts = count_image_keys(base_folder)
random_points,centers = generate_random_points(counts, r=10)
```

```python
random_points_input = np.array([list(tup) for tup in random_points])
centers_input = np.array([list(tup) for tup in centers])
start_time = time.time()
selected_greedy,queries,rewires,trace_curvature = greedy_coverage(centers_input, random_points_input, 10, 20)
duration = time.time() - start_time
#plot_environment(allocation_points, target_points, radius, selected)
print("Selected >> ",selected_greedy)
print(f"Total coverage for {50} selected points: {len(compute_coverage(centers_input[selected_greedy], random_points_input, 10))}")
print(f"Total time for {50} selected points: {duration/50}")
print(f"Total queries for {50} selected points: {queries}")
```

```python
random_points_input = np.array([list(tup) for tup in random_points])
centers_input = np.array([list(tup) for tup in centers])
start_time = time.time()
selected_resque,queries,rewires,trace_curvature = greedy_coverage(centers_input, random_points_input, 10, 20, "ResQue")
duration = time.time() - start_time
#plot_environment(allocation_points, target_points, radius, selected)
print("Selected >> ",selected_resque)
print(f"Total coverage for {50} selected points: {len(compute_coverage(centers_input[selected_resque], random_points_input, 10))}")
print(f"Total time for {50} selected points: {duration/50}")
print(f"Total queries for {50} selected points: {queries}")
```

```python
random_points_input = np.array([list(tup) for tup in random_points])
centers_input = np.array([list(tup) for tup in centers])
start_time = time.time()
selected_random,queries,rewires,trace_curvature = greedy_coverage(centers_input, random_points_input, 10, 20, "Random")
duration = time.time() - start_time
#plot_environment(allocation_points, target_points, radius, selected)
print("Selected >> ",selected_random)
print(f"Total coverage for {50} selected points: {len(compute_coverage(centers_input[selected_random], random_points_input, 10))}")
print(f"Total time for {50} selected points: {duration/50}")
print(f"Total queries for {50} selected points: {queries}")
```
